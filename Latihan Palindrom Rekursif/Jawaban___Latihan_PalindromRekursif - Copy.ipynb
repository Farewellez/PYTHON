````python
# Coding disini ya....
def is_palindrom(numbers):
    numbers = str(numbers)
    
    if len(numbers) <= 1:
        return True

    take_first = numbers[0]
    take_last = numbers[-1]

    if take_first != take_last:
        return False
    
    return is_palindrom(numbers[1:-1])

user_input = int(input("Input some number").lower().replace(" ",""))
print(is_palindrom(user_input))
````

### 7. Hanya Huruf
Deskripsi: <br>
Hilangkan semua karakter non-huruf sebelum mengecek palindrom.

Contoh Input:<br>
```
A man, a plan, a canal: Panama
```
Contoh Output: <br>
```
YES
```

````python
# Coding disini ya...
from string import ascii_letters as alpha

def is_palindrom(words):
    only_aplhabet = []
    
    for char in words:
        if char not in alpha:
            continue
        only_aplhabet.append(char)
    
    new_list = list(map(lambda x: x.lower(), only_aplhabet))

    if new_list != new_list[::-1]:
        return "No"
    return "YES"

words = "A man, a plan, a canal: Panama"
print(is_palindrom(words))
````

### 14. Dua Fungsi: Bersihkan dan Cek
Deskripsi: <br>

Contoh Input: <br>
```

```
Contoh Output: <br>
```

```

````python
# Coding disini ya...
import string

def clean_string(words):
    if not words:
        return ""
    
    first_words = words[0]

    if first_words in string.ascii_letters:
        return first_words.lower() + clean_string(words[1:])
    return clean_string(words[1:])

def is_palindrom(check):
    if len(check) <= 1:
        return True
    
    if check[0] != check[-1]:
        return False
    
    return is_palindrom(check[1:-1])
    

check_clean = clean_string("No 'x' in Nixon")
check_palindrom = is_palindrom(check_clean)

if check_palindrom == True:
    print("YES")
else:
    print("NO")
````

### 16. Efisien Tanpa Slicing
Deskripsi: <br>
Cek palindrom panjang (hingga 10⁴ karakter) tanpa slicing agar tidak boros memori.

Contoh Input:
```
"a" * 10000
```
Contoh Output:
```
YES
```

````python
# Coding disini ya...
def efficient_palindrom(words):
    start = 0
    last = len(words) - 1

    while start < last:
        if words[start] != words[last]:
            return False
        start += 1
        last -= 1
    
    return True

words = "a"*1000
print(efficient_palindrom(words))
````

### 17. Palindrom Terpanjang yang Bisa Dibentuk

```

# (Contoh palindrom: dccaccd)
```

````python
# Coding disini ya...
def count_palindrom(words):
    dict_word = {}
    odd_found = False

    for char in words:
        if char not in dict_word:
            dict_word[char] = 1
        else:
            dict_word[char] += 1
    
    count = 0
    for num in dict_word.values():
        if num % 2 != 0:
            odd_found = True
            count += num - 1
        else:
            count += num

    if odd_found == True:
        count += 1
    
    return count


words = [
    "abccccdd",
    "a",	
    "bb",	
    "abc",
    "aabbcc",	
    "aaaabbbb",	
    "aaabbb",	
    "abcd",
    "ccc",
    "cccccc",
    "abcdedcba",
    "xyzxyz",
    "xyzxy",	
    ]

for char in words:
    print(f"{char} = {count_palindrom(char)}")
````

### 19. Struktur Palindrom Tanda Kurung
Deskripsi: <br>
Cek apakah string tanda kurung merupakan palindrom secara struktur.

Contoh Input: <br>
```
({[]})
```
Contoh Output:
```
YES
```

````python
# Coding disini ya...
def is_palindrom(special):
    special_string = {
        "(": ")",
        "{": "}",
        "[": "]",
    }

    if len(special) % 2 != 0:
        return False

    left = 0
    right = len(special) - 1

    while left < right:
        if special_string[special[left]] != special[right]:
            return False
        left += 1
        right -= 1
    return True 

list_test = [
    "(]",
    "{(})",
    "[({})",
    "({[)]})",
    "((())",          
    "{[()]]}",        
    "([)]",           
    "({[]})",      
    "{[()]}", 
]
for char in list_test:
    print(is_palindrom(str(char)))
````

### 20. Palindrom dalam Inorder Binary Tree
Deskripsi: <br>
Diberikan binary tree, lakukan inorder traversal dan cek apakah hasilnya palindrom.

Contoh Input:
```
    a
   / \
  b   b
 /     \
a       a

```
Contoh Output:
```
YES
```

````python
# Coding disini ya...
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

root = Node("a")
root.left = Node("b")
root.right = Node("b")
root.left.left = Node("a")
root.right.right = Node("a")

result = []
def in_order(node, result):
    if node is None:
        return
    
    in_order(node.left, result)
    result.append(node.value)
    in_order(node.right, result)

in_order(root, result)

if result == result[::-1]:
    print(result)
    print("YES")
else:
    print(result)
    print("NO")
````

### 21. Palindrom DNA Mutan
Deskripsi: <br>
Sebuah perusahaan biotek memintamu menganalisis urutan DNA yang sangat panjang (hingga 10⁶ karakter). Mereka ingin mengetahui apakah terdapat urutan DNA yang merupakan palindrom sempurna dan tidak mengandung mutasi. Mutasi dilambangkan dengan karakter X, dan kamu harus mengabaikannya saat melakukan pengecekan.

Input:
```
"AXCGGTGCXCA"
```
Output:
```
NO
# (karena "ACGGTGCA" adalah palindrom setelah mengabaikan 'X')
```

````python
# Coding disini ya...
def dna_mutan(dna):
    # Hilangkan mutasi 'X' dan ubah ke huruf besar
    clean_dna = dna.replace("X", "").upper()

    # Dua pointer dari kiri dan kanan
    left = 0
    right = len(clean_dna) - 1

    while left < right:
        if clean_dna[left] != clean_dna[right]:
            return "NO"
        left += 1
        right -= 1

    return "YES"

# Contoh uji coba
print(dna_mutan("AXCGGTGCXCA"))   
print(dna_mutan("AXCGTGCXCA"))    
print(dna_mutan("XAXX"))          
print(dna_mutan("GGXGX"))         
print(dna_mutan("ACGTXGTCA"))     
````

### 23. Pemindai Palindrom Monumen Kuno
Deskripsi: <br>
Di sebuah situs arkeologi, ukiran batu yang sudah rusak berisi kata-kata dengan huruf hilang (dilambangkan _). Tentukan apakah mungkin mengisi huruf kosong dengan karakter apa saja agar menjadi palindrom. Gunakan rekursi untuk menjelajah semua kemungkinan.

Input:
```
r_c_r
```
Output:
```
YES
# (bisa menjadi racecar, recer, dll)
```

````python
# Coding disini ya...
from string import ascii_letters
from itertools import product

def old_monument(missing_words):
    missing = [i for i, char in enumerate(missing_words) if char == "_"]

    combinations = product(ascii_letters, repeat=missing_words.count("_"))

    all_possible = []
    for combination in combinations:
        missing_words = list(missing_words)

        for i, char in zip(missing, combination):
            missing_words[i] = char

        all_possible.append("".join(missing_words))
    return all_possible

words = "r_c_r"
result = old_monument(words)

def final_palindrom(result):
    final = set()
    
    for char in result:
        if char == char[::-1]:
            final.add(char)
    return list(final)

if final_palindrom(result):
    print("YES")
    print(final_palindrom(result)[:10])

````

### 24. Palindrom Zaman Mesir
Deskripsi: <br>
Hieroglif Mesir ditemukan berupa string karakter unik yang hanya bisa dibaca dari tengah ke luar (bukan dari ujung ke tengah). Buat fungsi rekursif untuk mengecek apakah string tersebut adalah middle-out palindrom (karakter di kiri dan kanan dari tengah harus sama).

Input:
```
racecar
```
Output:
```
YES
(karena cocok dari tengah ke luar: e == e, c == c, r == r)
```

````python
# Coding disini ya...
def hieroglif(words):
    if len(words) % 2 == 0:
        middle_toLeft = (len(words) // 2) - 1
        middle_toRight = len(words) // 2

        for _ in range(len(words) + 1):
            if middle_toLeft <= 0 or middle_toRight >= len(words):
                return True
            if words[middle_toLeft] != words[middle_toRight]:
                return False
            middle_toLeft -= 1
            middle_toRight += 1
    else:
        middle = len(words) // 2
        middle_toLeft = middle - 1
        middle_toRight = middle + 1

        for _ in range(len(words)):
            if middle_toLeft <= 0 or middle_toRight >= len(words):
                return True
            if words[middle_toLeft] != words[middle_toRight]:
                return False
            middle_toRight += 1
            middle_toLeft -= 1

test_cases = [
    "racecar",      # Palindrom ganjil karakter
    "level",        # Palindrom ganjil karakter
    "noon",         # Palindrom genap karakter  
    "redder",       # Palindrom genap karakter
    "a",            # Palindrom single karakter
    "",             # String kosong (dianggap palindrom)
    "hello",        # Bukan palindrom
    "python",       # Bukan palindrom
    "A man a plan a canal Panama",  # Palindrom dengan spasi dan kapitalisasi
    "No 'x' in Nixon",              # Palindrom dengan tanda baca
    "12321",        # Palindrom angka
    "12345",        # Bukan palindrom angka
    "Was it a car or a cat I saw?", # Palindrom kompleks
    "Madam, I'm Adam",              # Palindrom dengan tanda baca
    "race car",     # Palindrom dengan spasi
    "A Santa at NASA"  # Palindrom kreatif
]
for char in test_cases:
    print(f"{char} : {"YES" if hieroglif(char.replace(" ","").lower().replace("'","").replace("?","").replace(",","")) == True else "NO"}")
````

### 25. Neural Palindrome Decoder
Deskripsi: <br>
Sebuah model AI jaman purba hanya mengenali struktur nested palindrom. Sebuah string dikatakan nested palindrom jika string tersebut adalah palindrom dan bagian dalamnya (tanpa 1 huruf terluar di kiri dan kanan) juga merupakan nested palindrom. Buat fungsi rekursif untuk mengeceknya.

Input:
```
abccba
```
Output:
```
YES
# (karena: abccba → bccb → cc → "" [semua palindrom])
```

Input:
```
abcba
```
Output:
```
NO
# (karena: abcba → bcb → b [not nested palindrom, harus genap panjangnya])
```

````python
# Coding disini ya...
def old_ai(words):
    if len(words) <= 0:
        return True
    
    if len(words) % 2 != 0:
        return False
    
    if words[0] != words[-1]:
        return False
    return old_ai(words[1:-1])

test_cases = [
    "abccba",   # YES
    "abcba",    # NO
    "aabbcc",   # NO
    "aabbaa",   # YES
    "aa",       # YES
    "",         # YES
    "abcdba",   # NO
]

for char in test_cases:
    print(f"{char:<8}: {"YES" if old_ai(char) == True else "NO"}")

````
